```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#import dataset
ds = pd.read_csv('real_estate_dataset.csv')
ds
```

```python
ds.columns
```

```python
# assigning independent and dependent variables
x= ds[['Size_sqft']]
y=ds[['Price']]
```

```python
# plot the points
plt.scatter(x,y)
plt.xlabel('Size_sqft')
plt.ylabel('Price')
plt.show()
```

```python
# train test and fit dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
# apply linear regression  i.e fit LinearRegression on dataset
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
#predict result manually
result = reg.predict([[800]])
result
```

```python
res = reg.predict(x_test)
res
```

```python
y_test
```

```python
#draw the line to see the points are linear , scatter 
plt.scatter(x,y)
res= reg.predict(x)
plt.plot(x,res , color='red')
plt.show()
```

```python
#predict the result where input given by user
INR = int(input('Enter a sqft to predict its price : '))
myip = pd.DataFrame(columns=['Size_sqft'], data=[INR])
res = reg.predict(myip)
print(' price of flat per size of Sqft : ', round(res[0,0]))
```
